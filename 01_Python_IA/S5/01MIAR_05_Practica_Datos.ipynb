### Descarga de datos

```python
import requests
```

```python
url_park_valenvisi= 'https://valencia.opendatasoft.com/api/records/1.0/search/?dataset=valenbisi-disponibilitat-valenbisi-dsiponibilidad&q=&rows=100'

r = requests.get(url_park_valenvisi)
print(r) # si sale 200 es que va bien, si sale 400 se sale de contexto
```

```python
from datetime import datetime
```

```python
ls_estaciones= r.json()['records']
len(ls_estaciones)
ls_estaciones[0]
```

```python

def get_values(dc_station):
    updated = dc_station['fields']['updated_at']
    updated = datetime.strptime(updated, '%d/%m/%Y %H:%M:%S')
    adress = dc_station['fields']['address']
    number = dc_station['fields']['number']
    avaliable = dc_station['fields']['available']
    lat = dc_station['fields']['geo_point_2d'][0]
    lon = dc_station['fields']['geo_point_2d'][1]
    total =  dc_station['fields']['total']
    opend =  dc_station['fields']['open']
    free =  dc_station['fields']['free']
    return [updated, adress, number, avaliable, lat, lon, total, opend, free]

# ls_row = [get_values(station) for station in ls_estaciones]

ls_row = []
for station in ls_estaciones:
    ls_row.append(get_values(station))

ls_row[0]
```

```python
def estaciones_disponibles(tabla_valenbisi):

    avaliable = [n[3] for n in tabla_valenbisi]

    return sum(avaliable)

estaciones_disponibles(ls_row)
```

```python
class City:
    def __init__(self, city, state_short, state_full, country, alias):
        self.city = city
        self.state_short = state_short
        self.state_full = state_full
        self.country = country
        self.alias = alias
        

class CitiesDataRetriever():

    def __init__(self, url):
        self.cities = []
        self.url = url
        self.retrieve_data()

    def get_cities_by_state(self, state_short = None, state_full = None):
        if state_short:
            return [x.city for x in self.cities if x.state_short == state_short] # utiliza el propio atributo para hacer el filtro
        
        if state_full:
            return [x.city for x in self.cities if x.state_full == state_full]
    
    def count_cities_by_state(self, state_short = None, state_full = None):
        return len(set(self.get_cities_by_state(state_short, state_full)))
    
    def retrieve_data(self, separator= '|'):
        response = requests.get(self.url)

        if response.status_code == 200:
            data = response.text.strip().split('\n')
            for row in data:
                c = City(*row.split(separator))
                self.cities.append(c)

        else:
            print(f'Error: {response.status_code}')


source = CitiesDataRetriever(url= "https://raw.githubusercontent.com/grammakov/USA-cities-and-states/master/us_cities_states_counties.csv")
state_short = 'FL'
print(f'Number of cities in {state_short}: {source.count_cities_by_state(state_short=state_short)}')
```
