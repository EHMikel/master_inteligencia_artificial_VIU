# Lista de jugadores de futbol

Al finalizar la jornada 8 de una liga de futbol, se dispone en una lista (equipo) de listas (jugadores) con los registros de los jugadores. En cada registro se guarda el nombre del jugador, su edad y los minutos jugados en cada partido que haya jugado de estas jornadas. No todos los jugadores han jugado todos los partidos de estas 10 jornadas, por lo que en cada jugador aparecerán solo los minutos de los partidos jugados. Se muestra la lista del equipo (Usar col para las etiquetas de columns):

```python
lst_equipo = [['Lionel', 33, 77, 90, 82, 75, 90, 65, 90, 80], \
              ['Ramos', 34, 88, 75, 81, 90, 74, 62], \
              ['Koke',27, 78, 89, 85, 76, 90, 65, 81, 90], \
              ['Saul', 24, 90, 90, 90, 90, 90, 90, 90], \
              ['Joao', 25, 24, 22, 45, 20, 25, 90, 60], \
              ['Odriozola', 25, 70, 45, 90, 75, 80], \
              ['Lenglet', 23, 45, 55, 50], \
              ['Pique', 33, 90, 90, 90, 90, 90, 90, 90, 90], \
              ['Araujo', 21, 28, 15, 45, 18, 25, 33], \
              ['Valverde', 22, 45, 55, 33, 40], \
              ['Rosales', 30, 75], \
              ['Piatti', 40, 80, 31, 40, 20]]
```

```python
col = [ ('jornada'+ str(n)) for n in range(1,9)]
col.insert(0,'nombre')
col.insert(1,'edad')
col
```

```python
df4 = pd.DataFrame(lst_equipo, columns = col, index= range(len(lst_equipo)))
df4
```

a) Crear un DataFrame a partir de la lista de jugadores.A partir del DataFrame:

b) Calcula las series:

a. nombre de jugadores

b. suma de minutos jugados de cada jugador

c. promedio de los minutos jugados de cada jugador
c) Calcula el DataFrame con las series del apartado anterior. Resultado de la forma:

```python
df4['minutos_totales'] = None
df4['minutos_promedio'] = None

for n in df4.iterrows():
    
    df4.loc[n[0], 'minutos_totales'] = df4.loc[n[0], 'jornada1':'jornada8'].sum().round(2)
    df4.loc[n[0], 'minutos_promedio'] = df4.loc[n[0], 'jornada1':'jornada8'].mean().round(2)
    
df4
```

# Diccionario de vinos a DataFrame de vinos

Se dispone de un diccionario de vinos dV donde las claves son nombres de vinos y los valores para cada vino son una lista con la información de su denominación de origen, la añada y su precio.

```python
dV = {'Muga':['Rioja', 2012, 8.5],'Petit caus': ['Penedes',2014, 6.75],\
              'Viña vial': ['Rioja',2016,9.5], 'Lavia': ['Jumilla',2011,12.5], \
              'Bach': ['Penedes',2015, 4.5], 'Pazo B': ['Rias Baixas', 2016,13.0]}

colV = ['Nombre','Origen','Añada','Precio']
indV = ['V1','V2','V3','V4','V5','V6']
```

Crear un DataFrame a partir del diccionario dV que tenga en sus columnas 'Nombre', 'Origen', 'Añada', 'Precio' y los datos incluyan las claves y valores de las listas de vinos. Usar colV e indV para las etiquetas de columns e index.

```python
df5 = pd.DataFrame([(key, value[0],value[1], value[2]) for key, value in dV.items()],
                   columns= colV, index= indV)
df5
```

```python
# [(key, [n for n in value]) for key, value in dV.items()] # esto no es lo que quiero
```

```python
def parse_dV(dV): # devuelve una lista a partir de un diccionario que guarda listas en orden key, value[0], value[1], etc

    lV = []

    for key, value in dV.items():
        item = []

        item.append(key)
        for n in value:
            item.append(n)

        lV.append(item)

    return lV

# esto si es lo que quiero
```

```python
df6 = pd.DataFrame(parse_dV(dV), columns=colV, index= indV)
df6
```
